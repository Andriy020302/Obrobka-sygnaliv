The code loads an image and creates a picture collage consisting of 4 pictures. Each of the four pictures is a version of the original image with the colour channels swapped.

You will have to also finish some part of the code :-)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# Load an image (you can freely chose any image you like)
img = cv2.imread('kodim21.png') 
# Convert it to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Plot it
plt.imshow(img_rgb)
```

Let's make a collage.

```python
# Split the image into the three colour channels
red, green, blue = cv2.split(img)

# Compose the image in the RGB colour space
img1 = cv2.merge((red, green, blue))

# Compose the image in the RBG colour space
img2 = cv2.merge((red, blue, green))

# Compose the image in the GRB colour space
img3 = cv2.merge((green, red, blue))

# Compose the image in the BGR colour space
img4 = cv2.merge((blue, green, red))

# Create the collage
out1 = np.hstack([img1, img2])
out2 = np.hstack([img3, img4])
out = np.vstack([out1, out2])

# Plot the collage
plt.imshow(out)
plt.axis(False)
```

Conclusions:
1. При різних порядках каналів у зображення можуть відрізнятися кольори та відтінки. Наприклад, у зображенні з порядком каналів RGB (червоний, зелений, синій) червоний канал буде відповідати за червоні відтінки, зелений - за зелені, а синій - за сині. У зображенні з порядком каналів RBG (червоний, синій, зелений) порядок каналів змінюється, тому червоний канал відповідає за червоні відтінки, але зелений і синій поміняються місцями, що призведе до інших кольорів і відтінків у зображенні.

2. Це відбувається через те, що кольорова модель RGB визначає колір шляхом комбінування трьох основних кольорів - червоного, зеленого та синього - у різних пропорціях. Порядок цих каналів впливає на те, як кожен канал сприймається та інтерпретується.

3. Проблеми в додатках комп’ютерного зору, пов’язані з різним порядком колірних каналів, можуть включати невірну обробку кольору, помилкове розпізнавання об’єктів, які базуються на кольорі, та зміни відображення кольорів у зображеннях. Наприклад, якщо програма комп’ютерного зору очікує, що червоний канал буде завжди першим, але отримує зображення з порядком каналів RBG, це може призвести до неправильного аналізу чи обробки зображення.
